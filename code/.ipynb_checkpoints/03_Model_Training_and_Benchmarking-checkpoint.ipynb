# Model training and benchmarking

## Defining metrics for scoring models

Model evaluation is an important aspect of this project because it helps us understand the model's performance, and provides a fundamental basis for presenting the model to stakeholders/audience. While it is impossible to predict an exact value that later proves to be 100% correct, we actually seek to establish a set of metrics upon which we can measure how close our prediction is against the real value.

### R Square score

One of the metrics we would be using is the R Square score, which measures how much the variability in the dependent variable (i.e. targets, in this case the predicted property price) can be explained by the independent variable (i.e. predictors, in this case the set of property attributes). While the R Square score is a viable measure for judging how well the model fits the dependent variables, <b>it  does not account for cases of overfitting.</b> The latter effect is a syndrome of building a model with too many variables, resulting in it being overly complex. Such a model will predict well with seen data, but cope poorly with unseen data. From here on, we will refer to this metric as $R^2$ score.

$R^2$ score should generally fall between 0 and 1, but is possible to be negative too for cases where the error generated is so great, it makes the model worst off then simply providing a mean of seen targets as a naive prediction.

### Root-Mean Squared Error score

We begin by introducting Mean Squared Error, an absolute means to measure goodness of fit by computing the average of the summation of squared residuals. This metric would show how much the predcitions are deviating from the targets. The Root Mean Squared Error is simply the square-root of the mean Squared Error, is will be used in this project for the crucial reason that it projects the error in the same scale as the target, i.e. dollar ($). From here on, we will refer to this metric as RMSE score.

We calculate the RMSE score with known targets and predictions made from the model. There is no known range for RMSE score, but for model-to-model comparison, the better model is the one with a lower RMSE score (based on same set of predictor inputs). <b>For inter-model comparison, the better model has the lower RMSE score</b>.

For this project, <b>RMSE score on test-set would take greater weight in model evaluation, compared to $R^2$ score</b>.

### Root-Mean Squared Error Degradation score

For intra-model evaluation, we measure the gap between train-set RMSE error (seen data) and test-set RMSE error (unseen data). Assuming the abscence of poor sampling that leads to test data being bias against the train data in terms of distribution, we typically measure the degree of decline of testing RMSE score against training RMSE score. If the gap is huge, it indicates overfitting as the model has *memorised* the training-set, but unable to generalize to new data and just giving a higher RMSE score on the test-set. From here on, we will refer to this metric as RMSE Degradation score.

We calculate the RMSE Degradation score by taking the percentage increase in RMSE score for test-set, using the train-set RMSE score as a reference. For example if a model's train-set RMSE score is 10 and its test-set RMSE score is 11, the RMSE Degradation score would be 10%, as the test-set RMSE score is 1 unit higher than the train-set RMSE, and the percentage is calculated with train-set RMSE as the baseline. The closer the RMSE Degradation score to 0%, the better the model is at generalizing to new data. RMSE Degradation score is conventionally positive, as negative score would indicate that the model somehow did better on predicting the test data than the train data, of which we would need to re-evaluate the quality of our sampling for train/test data splitting.

For inter-model comparison, assuming their train-set RMSE scores are quite similar, the model with the lower RMSE Degradation score is the one that could generalize better to unseen data.

In other words, without reference to any of the features for a property targeted for price forecast, we simply provide a prediction value that is naively computed by averaging the known prices of all the properties in our train-set (seen data):

## Define score table for keeping track of model performance

```python
model_performance_df = pd.DataFrame(columns=['model_name', 'num_features', 'r2', 'train_rmse', 'test_rmse', 'pct_rmse_degradation'])
new_index = len(model_performance_df)
model_performance_df.loc[new_index, 'model_name'] = 'naive-means'
model_performance_df.loc[new_index, 'num_features'] = 0
model_performance_df.loc[new_index, 'r2'] = np.nan
model_performance_df.loc[new_index, 'train_rmse'] = float(naive_mean_train_rmse)
model_performance_df.loc[new_index, 'test_rmse'] = float(naive_mean_test_rmse)
model_performance_df.loc[new_index, 'pct_rmse_degradation'] = float(naive_mean_rmse_degradation)
model_performance_df
```

Going forward, every model that we have built for purpose of sale price prediction would be put into the above *score_table_df* for keeping track of their performance metrics. A name will also be given for each other model for easy reference. As a start, the model that provided the baseline score would be called *naive-means*.

```python
# define a helper function to keep track of model performance scores
def record_model_score(model_name, num_features, r2_score, train_rmse, test_rmse, rmse_degradation):
    if model_name in list(model_performance_df.loc[:, 'model_name']):
        new_index = model_performance_df.loc[model_performance_df.loc[:, 'model_name']==model_name, :].index
    else:
        new_index = len(model_performance_df)
    model_performance_df.loc[new_index, 'model_name'] = model_name
    model_performance_df.loc[new_index, 'num_features'] = num_features
    model_performance_df.loc[new_index, 'r2'] = r2_score
    model_performance_df.loc[new_index, 'train_rmse'] = float(train_rmse)
    model_performance_df.loc[new_index, 'test_rmse'] = float(test_rmse)
    model_performance_df.loc[new_index, 'pct_rmse_degradation'] = float(rmse_degradation)
```

## Training base model

We will train and benchmark the baseline model in the following steps:

1. Instantiate model
2. Perform 5-fold cross validation score to obtain a $R^2$ score, using training data from train_test_split
3. Fit model with training data from train_test_split
4. Compute RMSE (train) score using predictions made on train-set in step-3
5. Use model trained in step-3 to predict prices with test data from train_test_split
6. Compute RMSE (test) score using predictions made on test-set in step-4
7. Use train-set RMSE score from step-4 and test-set RMSE score from step-6 to compute RMSE Degradation score

### Define a function for model training and scoring

```python
def model_train_and_score(model_name, model, x_train_input, y_train_input, x_test_input, y_test_input, alpha=None):
    result = {}
    # 1. instantiate model
    # *done during function call*
    
    if type(model) not in [RidgeCV, LassoCV, ElasticNetCV]:
        # 2. perform 5-fold cross-validation to obtain r-squared score, using training data from train_test_split
        r2_score = cross_val_score(estimator=model, X=x_train_input, y=y_train_input, cv=5).mean()
        print('{} model train-set R2 score: {}'.format(model_name, np.round(r2_score, 4)))
    
    # 3. fit model with train data, obtain RMSE error
    model.fit(X=x_train_input, y=y_train_input)
    
    # 4. use model trained in step-3 to predict prices with train data, obtain RMSE error
    training_rmse = np.sqrt(mean_squared_error(
        y_train_input, 
        model.predict(x_train_input)))
    print('{} model train-set RMSE score: {}'.format(model_name, np.round(training_rmse, 4)))
    
    # 5. use model trained in step-3 to predict prices with test data
    y_test_pred = model.predict(x_test_input)
    
    # 6. compute RMSE score using test data
    testing_rmse = np.sqrt(mean_squared_error(y_test_input, y_test_pred))
    print('{} model test-set RMSE score: {}'.format(model_name, np.round(testing_rmse, 4)))
    
    #7. compute degradation score
    rmse_degradation = np.round((testing_rmse/training_rmse - 1)*100, 2)
    print('RMSE degradation score: {}%'.format(rmse_degradation))
    
    result['model_name'] = model_name
    result['num_features'] = x_train_input.shape[1]
    result['model'] = model # trained model
    result['r2'] = r2_score
    result['training_rmse'] = training_rmse
    result['testing_rmse'] = testing_rmse
    result['rmse_degradation'] = rmse_degradation
    
    return result
```

Next we take inspect at the coefficients in *LR-1*. 

Training data fed into *LR-1* (and subsequent models built here) has been standardized to mean of 0 and standard deviation of 1. As a consequence, the fitted regression model would now have standardized coefficients too, which are not directly interpretable as dollar-based effects on the sale price. However, like the standardized predictor values, the coefficients are now on similar scale to each other and we would be able analysis their relationship as relative strength to each other.

```python
# helper method to facilitate display of dataframes side by side
def combine_display(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html() # naive concat the html content together
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
def generate_coefficient_tables(model, predictor_columns, scaler, target_stdev_value, 
                                cutoff=10, is_show_bottom_coefs=False, is_show_stddev_statistics=False):
    # retrieve coefficients from trained linear model (baseline)
    if type(model) in [Lasso, ElasticNet]:
        coefficients = model.coef_
    else:
        coefficients = model.coef_[0]
    
    # build dataframe of column names and their respective coefficient values
    if is_show_stddev_statistics:
        standardized_coefficient_df = pd.DataFrame({
            'predictor': predictor_columns,
            'coefficient': coefficients,
            'Predictor Std. Dev.': np.sqrt(scaler.var_)
        })
        standardized_coefficient_df.loc[:,'Change in SalePrice due to 1 stdev. change in predictor'] = \
                                standardized_coefficient_df.loc[:,'coefficient'] * target_stdev_value
    else:
        standardized_coefficient_df = pd.DataFrame({
            'predictor': predictor_columns,
            'coefficient': coefficients
        })

    # create dataframe of top 10 highest coefficients
    top_predictors = standardized_coefficient_df.loc[:,'coefficient'].abs().sort_values(ascending=False).head(cutoff).index
    top_coef_df = standardized_coefficient_df.loc[top_predictors,:]#.reset_index(1, drop=True)
    top_coef_df.index = np.arange(1, len(top_coef_df)+1)
    top_coef_df.index.name = 'Top {} strongest coefs.'.format(cutoff)

    # create dataframe of top 10 lowest coefficients
    bottom_predictors = standardized_coefficient_df.loc[:,'coefficient'].abs().sort_values(ascending=True).head(cutoff).index
    bottom_coef_df = standardized_coefficient_df.loc[bottom_predictors,:]#.reset_index(1, drop=True)
    bottom_coef_df.index = np.arange(1, len(bottom_coef_df)+1)
    bottom_coef_df.index.name = 'Top {} weakest coefs.'.format(cutoff)

    if is_show_bottom_coefs:
        combine_display(top_coef_df, bottom_coef_df)
    else:
        combine_display(top_coef_df)
    #print('\ny_train standard deviation: {}'.format(target_stdev_value))
    
    return standardized_coefficient_df
```

```python
lr_1_coef_df = generate_coefficient_tables(result['model'], 
                                           [col for col in x_train_baseline.columns], 
                                           ss,
                                           np.std(y_train_baseline.loc[:,'SalePrice']))
```

In the above table of the top 10 standardized coefficients rated by their absolute magnitude, we observe unsual comparative strengths. In the earlier *Feature Engineering* phase, both `MS SubClass_C090` and `Bldg Type_Duplex` were not within the top 30 most correlated variables with `SalePrice`, yet demonstrated extremely high strength in influencing the predictions in *LR-1*, compared to other variables such as `Gr Liv Area` and `Overall Qual`. These overblown coefficients have led us to believe that *LR-1* is not a viable model as those unexplainably strong coefficients is likely a symptom of overfitting with too many predictors.

## Ridge Model with 40 predictors

At this point, we would like to explore the possiblity of reducing the number of predictors further, to see if it can help the  linear model generalize better. The next model to attempt is the Ridge model. It is a form of linear model, with the additional feature of accepting a regularization parameter (called alpha) which would impost a penalty component on the model coefficients using a loss function.

Before we proceed with an actual Ridge model, we will first use a RidgeCV model. This RidgeCV model essentially runs Ridge mechanics over a given range of alpha values. The aim of RidgeCV is to use cross-validation internally to derive an optimal alpha value that would give the best R Squared score. With an optimal alpha score, we then perform the actual step of training a Ridge model that would perform the actual prediction.

First, we instantiate a RidgeCV model with a given range of alpha values:

```python
ridge_alpha_values = np.logspace(0, 5, 200)
ridge_cv_model = RidgeCV(ridge_alpha_values, store_cv_values=True)
```

```python
if type(ridge_cv_model) in [RidgeCV, LassoCV, ElasticNetCV]:
    print('True')
```

We then use the *.fit* function call to trigger the cross-validation internally over the given range of alpha values:

```python
ridge_cv_model.fit(X=x_train2, y=y_train2)
```

We can extract the optimal alpha value in the following manner:

```python
ridge_cv_optimal_alpha = ridge_cv_model.alpha_
print('RidgeCV optimal alpha: {}'.format(ridge_cv_optimal_alpha))
```

Next, we train the Ridge model with the optimal alpha value from above. This is the model that we would train and evaluate, and we will call it *Ridge-1*.

```python
result_ridge_1=model_train_and_score('Ridge-1',
                                         Ridge(alpha=ridge_cv_optimal_alpha),
                                         x_train2,
                                         y_train2,
                                         x_test2,
                                         y_test2)

record_model_score(
    result_ridge_1['model_name'], 
    result_ridge_1['num_features'], 
    result_ridge_1['r2'], 
    result_ridge_1['training_rmse'], 
    result_ridge_1['testing_rmse'], 
    result_ridge_1['rmse_degradation'])

model_performance_df
```

As shown above, test-set RMSE score and RMSE Degradation score has improved slightly over *LR-2*, which gives an indication that regularization has helped to dampen the coefficients of some of the predictors to generalize the model slightly.

## Lasso model with 40 predictors

Another linear model with regularization feature is the Lasso model. The difference is that in Ridge, the coefficients are subjected to a regularization parameter (also called alpha) that would bring some closer to zero, while in Lasso, the regularization parameter has the capability to make some coefficients reach 0. This *zeroing* effect is the equivalent of indicating which predictors can be taken out of the model.

However, like Ridge model, we have to first get an optimal alpha value, and this can be done with LassoCV. As the name suggest, it would use cross-validation internally to get an alpha value that would yield the best R Square score, from which we can use in an actual Lasso model.

First of all, we instantiate a LassoCV model with an alpha count of 100. This time, we indicate the number of alphas to cross-validate, and the LassoCV model would internally generate a range of that given quantity to search upon:

```python
alpha_count=100
lasso_cv_model = LassoCV(n_alphas=alpha_count)
```

Like RidgeCV, we then use the *.fit* function call to trigger the cross-validation internally over the given number of alpha values:

```python
lasso_cv_model.fit(X=x_train2, y=y_train2.loc[:,'SalePrice'])
```

Optimal alpha value is also extract in a similar manner:

```python
lasso_cv_optimal_alpha = lasso_cv_model.alpha_
print('LassoCV optimal alpha: {}'.format(lasso_cv_optimal_alpha))
```

Next, we train the Ridge model with the optimal alpha value from above. This is the model that we would train and evaluate, and we will call it *Lasso-1*.

```python
result_lasso_1=model_train_and_score('Lasso-1'.format(lasso_cv_optimal_alpha),
                                 Lasso(alpha=lasso_cv_optimal_alpha),
                                 x_train2,
                                 y_train2,
                                 x_test2,
                                 y_test2)

record_model_score(
    result_lasso_1['model_name'], 
    result_lasso_1['num_features'], 
    result_lasso_1['r2'], 
    result_lasso_1['training_rmse'], 
    result_lasso_1['testing_rmse'], 
    result_lasso_1['rmse_degradation'])

model_performance_df
```

*Lasso-1* seems to be performing better than *Ridge-1*, with test-set RMSE score and RMSE degradation score improving slightly more when compared with *LR-2*. While all their test-set RMSE scores still cannot compare with *LR-1*, the subsequent models have shown continuous small improvements in generalizing better.

## Backward Elimination using Ordinary Least Squares

One possible method of reducing predictors in a model is the use of Ordinary Least Squares (OLS) from the statsmodels.api package. It is another form of linear regression model, with the added feature that it would generate the p-value for the null hypothesis that the given feature has significant impact on the prediction output. For p-value above a specified significance level (usually 0.05, or 5%), we will reject the null hypothesis, i.e. there is insufficient evidence that feature has significant impact, thus justifying its removal.

This is performed in an iterative manner, in which at every cycle, we will extract the feature with highest p-value after performing the training of the OLS model. If p-value is above 0.05, we remove it from the set of predictors and repeat the OLS process. If no features are found with p>0.05, the whole iterative mechanism is stopped. Hence the name *Backward Elimination*, where the set of iterators is progressively trimmed down.

```python
x_train_reduced_after_lasso_reduction = x_train_reduced_after_lasso1.copy()

# capture current list of columns (should be 37 after the lasso reduction)
variables_before_BE = list(x_train_reduced_after_lasso_reduction.columns)
num_features_before_BE = len(variables_before_BE)
pmax = 1
print('Before backward elimination: {}\n'.format(variables_before_BE))
iteration_count = 1

# while there are still variables to trim down
while (len(variables_before_BE)>0):
    p_values= []
    X_1 = x_train_reduced_after_lasso_reduction.loc[:, variables_before_BE]
    X_1 = sm.add_constant(X_1)
    
    # fit the OLS model
    model = sm.OLS(y_train_reduced_and_engineered, X_1).fit()
    
    #extract all available p-values
    p_values = pd.Series(model.pvalues.values[1:], index = variables_before_BE)
    
    #find maximum p-value amongst all the variables
    pmax = max(p_values)
    
    #extract corresponding variable for the largest p-value
    feature_with_p_max = p_values.idxmax()
    if(pmax>0.05):
        print('Iteration #{}, removing `{}` with p-value={}'.format(iteration_count, feature_with_p_max, pmax))
        #remove feature from list of variables, then repeat backward elimination with new iteration
        variables_before_BE.remove(feature_with_p_max)
        iteration_count += 1
    else:
        # if the largest p-value does not surpass 0.05, terminate the whole backward elimination process
        break

variables_after_BE = variables_before_BE
print('\nBefore backward elimination: {}\n'.format(variables_after_BE))

num_variables_removed = num_features_before_BE - len(variables_after_BE)

print('\nNumber of variables removed: {}\nNumber of remaining variables: {}'.format(num_variables_removed, len(variables_after_BE)))
```

Backward Elimination has managed to detect and remove 13 variables that are deemed to be of no significant impact to model prediction of prices. We now make the necessary revisions to the train and test datasets:

```python
ss_x_train_reduced_after_BE = StandardScaler()
x_train_reduced_after_BE_scaled = ss_x_train_reduced_after_BE.fit_transform(x_train_reduced_after_lasso_reduction.loc[:, variables_after_BE])

x_train4, x_test4, y_train4, y_test4 = train_test_split(
    x_train_reduced_after_BE_scaled, 
    y_train_reduced_and_engineered, 
    test_size=0.2, 
    random_state=1)
print('x_train4 shape: {}'.format(x_train4.shape))
print('x_test4 shape: {}'.format(x_test4.shape))
print('y_train4 shape: {}'.format(y_train4.shape))
print('y_test4 shape: {}'.format(y_test4.shape))
```

## ElasticNet Model with 24 predictors

At this juncture, we attempt to build an ElasticNet model for sale price prediction. ElasticNet is a linear model that is capable of combining both Ridge and Lasso penalties to the loss function that is dampening the coefficients. We have to provide the *alpha* value, as well as the *l1-ratio* value to determine the level of effect between the Ridge and Lasso mechanisms.

Again, in order to arrive at the optimal *alpha* and *l1-ratio* values, we will resort to an ElasticNetCV to perform the search via cross-validation to determine the best combination of parameters that would give the best R Squared score:

```python
# Define a helper function to perform the search over a given l1_ratio
def elasticnetcv_search(x_input, y_input, n_alpha_value, l1_ratio_range, max_iterations=1000):
    result = {'alpha': [], 'l1_ratio': [], 'r2_score': []}
    enet_model = ElasticNetCV()
    for l1_ratio_value in l1_ratio_range:
        enet_model.set_params(n_alphas=n_alpha_value, l1_ratio=np.round(l1_ratio_value, 1), max_iter=max_iterations)
        r_squared_score = cross_val_score(estimator=enet_model, X=x_input, y=y_input, cv=5).mean()
        enet_model.fit(x_input, y_input)
        #coefs.append(enet_model.coef_)
        result['alpha'].append(enet_model.alpha_)
        result['l1_ratio'].append(l1_ratio_value)
        result['r2_score'].append(r_squared_score)
    return result
```

After defining the helper function to execute the ElasticNetCV search logic, we proceed to build the ElasticNetCV with an alpha count of 100, with a ratio range spanning 0.1 to 1.0:

```python
alpha_count=100
l1_range = np.arange(0.1, 1.1, 0.1)
max_iterations=2500
en_cv_result = elasticnetcv_search(x_train4, y_train4.loc[:,'SalePrice'], alpha_count, l1_range, max_iterations)
```

After completion of the ElasticNetCV search, we inspect the search results:

```python
en_cv_result_df = pd.DataFrame(en_cv_result)
en_cv_result_df
```

Turns out that the ElasticNetCV is in favor of a 100% lasso-style ElasticNet model, as given by the set of parameters for the highst R Squared score of 0.88174. We then proceed to build the ElasticNet model with those optimal parameters, naming it *ElasticNet-1*:

```python
# capture optimal l1-ratio and alpha
selected_en_l1_ratio = en_cv_result_df.loc[en_cv_result_df.loc[:,'r2_score'].idxmax(),'l1_ratio']
selected_en_alpha = en_cv_result_df.loc[en_cv_result_df.loc[:,'r2_score'].idxmax(),'alpha']

print('Optimal l1-ratio: {}'.format(selected_en_l1_ratio))
print('Optimal alpha: {}'.format(selected_en_alpha))
```

```python
result_en_1=model_train_and_score('ElasticNet-1',
                                 ElasticNet(alpha=selected_en_alpha, l1_ratio=selected_en_l1_ratio),
                                 x_train4,
                                 y_train4,
                                 x_test4,
                                 y_test4)

record_model_score(
    result_en_1['model_name'], 
    result_en_1['num_features'], 
    result_en_1['r2'], 
    result_en_1['training_rmse'], 
    result_en_1['testing_rmse'], 
    result_en_1['rmse_degradation'])

model_performance_df
```
